```python
product_name=[]
price_list=[]
reviews_list=[]
```

```python
# we scrap only one page , other response denied
# for i in range(2,12): 
url = "https://www.flipkart.com/search?q=mobiles+under+30000+5g&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_13_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_13_na_na_na&as-pos=1&as-type=RECENT&suggestionId=mobiles+under+30000+5g%7CMobiles&requestId=380e7024-0cb1-4e99-84ee-2a336dac8183&as-searchtext=mobiles+under&page="+str(1)

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

box = soup.find("div", class_ = "DOjaWF gdgoEp")  # to not add other mobile reviews which are not in list

names = box.find_all("div",class_ = "KzDlHZ")

for i in names:
    name = i.text
    product_name.append(name)

# print(product_name)

prices = box.find_all("div",class_ = "Nx9bqj _4b5DiR")

for i in prices:
    price = i.text
    price_list.append(price)

# print(price_list)

reviews = box.find_all("div",class_ = "XQDdHH")

for i in reviews:
    review = i.text
    reviews_list.append(review)
# print(len(reviews_list))
```

```python
df = pd.DataFrame({"Product Name":product_name, "Price": price_list, "Review":reviews_list})
```
